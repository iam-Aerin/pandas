```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
df.head()
```

## 컬럼 추가

```python
df_copy = df.copy()
```

```python
df_copy['VIP'] = False
#일괄적으로 똑같은 값을 넣을때
```

```python
df_copy.head(2)
```

```python
#가지고 있는 column들을 연산해서 => 새로운 column을 생성 할 수 있다. 
df_copy['family'] = df_copy['sibsp'] + df_copy['parch']
df_copy.head(2)
```

```python
#숫자끼리의 연산이 아닌, 글자 연산으로도 새로운 column을 만들어낼 수 있다.
df_copy['gender'] = df_copy['sex']+ '-' + df_copy['who']
df_copy.head(2)
```

```python
#나이 당 얼마의 요금을 냈는지를 알아보기위해
#나누기 연산이 들어간 조금 더 복잡한 연산이 들어간 새로운 column 만들기
df_copy['f/a'] = round(df_copy['fare'] / df_copy['age'], 2)
#round(수식, n): 소수점 두번째 자리까지 남기고 반올림.
df_copy.head(1)
#특성공합
```

## cut()
> 내가 일정 비율로 데이터를 나눠서 볼 지/ 수치로 나눠서 볼 지를 선택 할 수 있는 함수: cut()

```python
bins = [0, 100, 200, 300, df_copy['운동량'].max()]
#list리스트의 범위를 먼저 지정해서 설정해준다. 
pd.cut(df_copy['운동량'], bins)
# pd.cut(내가 원하는 데이터의 column, 구간 리스트)
```

```python
pd.cut(df_copy['운동량'], bins=10)
#데이터의 구간을 10으로 나눔 (하지만, 데이터 내에 좋지 못한 데이터: 너무 작거나/ 너무 큰 )가 포함되어 있다면, 별로인 방법)
pd.cut(df_copy['운동량'], bins=10).value_counts()
```

```python
#qcut: 비율대로 나누는 함수
pd.qcut(df_copy['운동량'], q=10)
pd.qcut(df_copy['운동량'], q=10).value_counts()
```

```python
#qcut으로 비율을 나눌 때 표준분포 처럼 나누고 싶다면
qcut_bins = [0, 0.2, 0.8, 1]
df_copy['운동량_한글'] = pd.qcut(df_copy['운동량'], qcut_bins, labels=['적음', '보통', '많음'])
#labels=[]로 라벨링을 달아줄 수 있음.
df_copy.head()
# (93.414, 3328.186]         196098 중간에 있는 60%의 사람이다.
```
