## fillna()

```python
df_copy = df.copy()
```

```python
df_copy.tail()
```

```python
#fillna()함수로 결측치에 어떠한 값을 넣어주세요~
df_copy['age'].fillna(100)
```

```python
df_copy.tail()
#fillna()는 연산결과만 반영하고, 실제 데이터프레임을 수정하지는 않는다.
```

```python
# 만약 copy 버전, 그 자체를 수정하고 싶다면
df_copy['age'] = df_copy['age'].fillna(100)
# 처럼 그 데이터프레임에 재할당을 해줘야 한다. 
df_copy.tail()
```

```python
df_copy.info()
#info()함수는 내가 가진 데이터의 구조를 알려줌. 
```

```python
df_copy['deck'].value_counts()
#df 중 'deck' column 정보를 확인했다. 
```

```python
# df_copy['deck'].fillna('C')
# nan 을 위에서 가장 많은 것으로 나타난 'C'로 대입해줌. 
# df_copy['deck'].fillna('Z')
# 내가 category로 지정하지 않은 데이터값은 넣을 수가 없음. => 새로운 category로 넣어줘야 사용이 가능하다.
```

```python
df_copy['deck'] = df_copy['deck'].cat.add_categories('Z')
df_copy['deck'].fillna('Z')
#일괄적으로 하는 것 -> 좋지는 않은데, 일단은 할 수 있다는 걸 보려고 하는 중이여요~
#제대로하려면 => 통계적인 수치를 계산해서 결측치를 채우는 방법이 더 좋아요~
#그 방법은 아래에서 계속됩니다~
```

```python
age_mean = df_copy['age'].mean()
df_copy['age'] = df_copy['age'].fillna(age_mean)
#'age'의 결측값을 'age의 평균값'으로 채워넣어
```

```python
df_copy['age'] = df_copy['age'].fillna(age_mean)
#'age'의 결측값을 'age의 평균값'으로 채워넣어
```

## practice. 예제 문제

- 남자 승객의 결측치는 남자 승객의 평균으로
- 여자 승객의 결측치는 여자 승객의 평균으로
- 결측치를 모두 채우고 나서 모든 사람의 평균으로 

```python
df_copy = df.copy()
```

```python
df_copy.head()
```

```python
df_copy.isna().sum()
```

```python
## 결측치가 age에서 확인되므로
##남자의 결측치 age는 남자의 age평균으로
##여자의 결측치 age는 여자의 age평균으로 
#채워준다. 
```

```python
cond1 = df_copy['sex'] == 'male'
male_mean = df_copy.loc[cond1, 'age'].mean()
print(male_mean)
```

```python
cond2 = df_copy['sex'] == 'female'
female_mean = df_copy.loc[cond2, 'age'].mean()
print(female_mean)
```

```python
df_copy.loc[cond1, 'age'] = df_copy.loc[cond1, 'age'].fillna(male_mean)
df_copy.loc[cond2, 'age'] = df_copy.loc[cond2, 'age'].fillna(female_mean)
```

```python
df_copy['age'].mean()
```
