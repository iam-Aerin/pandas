## practice (예제)

```python
tips = sns.load_dataset('tips')
tips
# seaborn 안의 'tips' 라는 데이터를 불러와서 연습하기.
```

- 남성 흡연자 조건 필터링
- tip을 가장 많이 낸 순서로 내림차순 정렬
- 상위 10개 출력

```python
#남성이면서 흡연자인 경우를 먼저 불러오자

cond1 = tips['smoker'] == 'Yes'
cond2 = tips['sex'] == 'Male'

# # cond = tips.loc[cond1 & cond2].head(10)
# # # & 기호를 통해 
# cond.sort_values('tip', ascending=False).head(10)

tips.loc[cond1 & cond2].sort_values('tip', ascending=False).head(10)
```

- 식사인원이 5명 이상인 저녁테이블
- 흡연자가 없는 주말 테이블
- total_bil, tip 출력

```python
# time == Dinner
# size == 5 
# smoker == No
# day == Sat | Sun
# total_bill, tip

cond1 = tips['time'] ==  'Dinner'
cond2 = tips['size'] >= 5

cond3 = tips['smoker'] == 'No'
cond4 = tips['day'].isin(['Sat', 'Sun'])


# tips.loc[(cond1 & cond2) & (cond3 & cond4)][['total_bill', 'tip']]
tips.loc[(cond1 & cond2) & (cond3 & cond4), ['total_bill', 'tip']]
```
