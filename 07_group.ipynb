# Groupby
> 똑같은 column들을 묶어서 보는 개념

```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head(3)
```

```python
# 성별을 기준으로 나눠보겠다.
# 남자 생존율의 평균/ 여자 생존율의 평균값을 보고 싶어요.
df_copy = df.copy()
```

## groupby()
> 내가 지금 보려는 데이터를 누구를 기준으로 묶어줄까?
>  하나의 column을 기준으로 데이터를 묶어서 본다.


```python
df_copy.groupby('sex').mean(numeric_only=True)
```

```python
df_copy.groupby(['sex', 'pclass']).mean(numeric_only=True)
#groupby([리스트로 여러개의 column을 넣어줄 수 있음.])
#groupby: 앞에서부터 데이터를 묶어준다. 
```

```python
#생존율이 궁금한거기 때문에
# df_copy.groupby(['sex', 'pclass'])['survived'].mean(numeric_only=True)
#[survived]:단일 데이터에 접근하는 것이므로, Series로 읽어
df_copy.groupby(['sex', 'pclass'])[['survived']].mean(numeric_only=True)
#[[survived]]: 데이터프레임으로 보여줘 - 더 보기 좋게 나옴.
```

## reset_index()

```python
df_temp = df_copy.groupby(['sex', 'pclass']).mean(numeric_only=True)
```

```python
df_temp.reset_index()
#다시 groupby한 내용을 초기화
#여기 부분 코드가 맞는지 확인 해야 함.
```

```python
df_copy.groupby(['sex', 'pclass'])[['survived', 'age']].agg(['mean', 'sum'])
```

## pivot_table
> 사실상 groupby()랑 거의 비슷하다.

```python
df_copy.pivot_table(index='who', values='survived')
#각각의 column이 어떠한 관계를 가지고 있는지를 표시해줌.
```

```python
#여러개의 table도 구성이 가능하다. 
df_copy.pivot_table(index=['who', 'pclass'], values='survived')
```
