## concat()
> 리스트로 넣어 (2개 이상 가능하다)

```python
pd.concat([df1, df2])
#index가 중복되어서 찍힌 것을 볼 수 있음 => 마지막240row의 index가 57인 것을 볼 수 있기에
```

```python
# pd.concat([df1, df2]).iloc[170:199] 
#연결점이 어떤지를 확인하려고 위 코드로 확인함.
pd.concat([df1, df2], ignore_index=True)
```

```python
# 열방향으로 concat 하려면
sales1 = df1.iloc[:, :5]
# loc[위에서부터 아래까지의 데이터 전체, :컬럼을 의미-첫번째 부터 5번째 column까지를 의미]
sales2 = df2.iloc[:, 5:]
#각각의 column들을 `좌우로` 이어보려고 
#만약 iloc가 아닌 loc로 하려면, 인덱스를 직접 적어줘야한다. 
```

```python
sales1.info()
```

```python
sales2.info()
```

```python
pd.concat([sales1, sales2], axis=1)
#axis=1을 적음으로써 좌우로 합치겠다. 
```

## merge()
> 병합

```python
info = {
    '고객명': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']
}

money = {
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']
}
```

```python
#지금가지고 있는 데이터가 dictionary이므로, 이걸 우선 data frame으로 바꿔주자.
df1 = pd.DataFrame(info)
df2 = pd.DataFrame(money)
```

```python
df1
```

```python
df2
```

```python
#data frame을 확인했을때, 중요한점이 지금 -> 중복된 이름과 중복되지 않은 이름이 발견된다. 
#단일 데이터가 원자로써 존재해야한다.
#독립성
#데이터는 한 칸에 하나만 있게끔
```

```python
# pd.merge(df1, df2)
# ---
#마지막에 있던 마이클조던, 타이거우즈가 사라짐
#교집합만 남아있게된다: merge 
#join과 동일한 개념 - SQL JOINS 
#교집합 : INNER JOIN
#LEFT JOIN / RIGHT JOIN / OUTER JOIN 

pd.merge(df1, df2, how='inner')
#how를 통해 option옵션을 지정해 줄 수 있다. 
#inner: 교집합
```

```python
pd.merge(df1, df2, how='outer')
#전체데이터
#비어있는 값은 NaN 
```

```python
pd.merge(df1, df2, how='right')
```

```python
pd.merge(df1, df2, how='left')
```

```python
#만약 데이터가 다르다면?
#고객명 - 고객명 상태가 아니라 고객명 - 이름 의 데이터프레임 상태라면?

info = {
    '이름': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']
}

money = {
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']
}
```

```python
df1 = pd.DataFrame(info)
df2 = pd.DataFrame(money)
```

```python
pd.merge(df1, df2, left_on='이름', right_on='고객명')
#내가 직접적으로 기준이 되는 column의 이름을 적어주는 left_on, right_on
```

```python
movies = pd.read_csv('data/movies.csv')
reviews = pd.read_csv('data/reviews.csv')
```

```python
movies.head(2)
```

```python
reviews.head(2)
```

```python
#2018년에 개봉한 영화의 모든 review를 보고 싶다면?
#2018년 -> movies에서 / review -> reviews에서 가져와야하므로
#merge 해야한다.
movie_review =  pd.merge(movies, reviews)
```

```python
# movie_review.head(5)
#잘 merge됐는지를 확인해봄
```

```python
cond = movie_review['movieYear'] == 2005
movie_review.loc[cond, 'rating'].mean()
#2005년에 개봉한 영화들을 먼저 필터링 하고, rating (평점)의 평균을 구함
```
